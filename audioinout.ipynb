Read file in with song=wave.open then save its parameters to be copied later. IPAUDIO displays the audio file. Only wav files work

```python
import wave
from IPython.display import Audio as IPAudio
song=wave.open("Communication Breakdown (Remaster).wav")
parameters = song.getparams()
frames = song.readframes(song.getnframes())
IPAudio("Communication Breakdown (Remaster).wav")
```

output is a new audio file that is being written to (wb) and shares the frames and parameters of the original we can listen to our output file again with IPaudio.

```python
output=wave.open("output.wav","wb")
output.setparams(parameters)
output.writeframes(frames)
IPAudio("output.wav")
```
