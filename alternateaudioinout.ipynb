```python
import requests
import os
import json
import librosa
import soundfile
import ipywidgets as widgets
from IPython.display import display, clear_output, Markdown

button = widgets.Button(description="Search")
query_box = widgets.Text(value="daft punk")

output = widgets.Output()

def SearchDeezer(query):
    url = f"https://api.deezer.com/search?q={query}"
    response = requests.get(url)
    data = response.json()
    tracks = data["data"]



    print("Song Selected: ", tracks[0]["title"])
    print()

    preview_url = tracks[0]["preview"]
    audio = requests.get(preview_url)

    with open("preview.mp3", "wb") as f:
        f.write(audio.content)

    print("Downloaded mp3")
    y, sr = librosa.load("preview.mp3", sr=None)
    soundfile.write("preview.wav",y,sr)
    print("Downloaded WAV")
    

def RunSearch(b):
    with output:
        clear_output()
        results = SearchDeezer(query_box.value)


display(Markdown("""Search for a song"""))
button.on_click(RunSearch)
display(query_box, button, output)
```
